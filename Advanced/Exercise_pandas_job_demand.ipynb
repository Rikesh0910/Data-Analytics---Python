```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")

df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_US = df[df["job_country"] == "United States"].copy()
df_US
```

```python
df_US["job_posted_month"] = df_US["job_posted_date"].dt.strftime("%B")
df_US
```

```python
pivot_df_US = df_US.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")

pivot_df_US

```

```python
pivot_df_US.reset_index(inplace=True)

pivot_df_US["month_no"] = pd.to_datetime(pivot_df_US["job_posted_month"], format="%B").dt.month

pivot_df_US.sort_values("month_no", inplace=True)

pivot_df_US.set_index("job_posted_month", inplace=True)

pivot_df_US.drop(columns="month_no", inplace=True)
```

```python
pivot_df_US.drop(columns="index", inplace=True)
pivot_df_US
```

```python
top_3 = df_US["job_title_short"].value_counts().head(3)

top_3 = top_3.index.to_list()

top_3
```

```python
pivot_df_US[top_3].plot(kind="line")
plt.title("Top data jobs for jobs in US")
plt.ylabel("job count")
plt.xlabel("2023")
plt.legend()
plt.show()
```
