```python
from datasets import load_dataset

import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")

df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df.info()
```

```python
df["job_title_short"].value_counts().sort_values(ascending=True)
```

```python
df.index.name = "job_index"

df.head(6)
```

```python
df_usa_jobs = df[df["job_country"] == "United States"]

df_usa_jobs
```

```python
median_pivot = df_usa_jobs.pivot_table(values="salary_year_avg", index="job_title_short", aggfunc=["median", "max", "min"])
median_pivot
```
