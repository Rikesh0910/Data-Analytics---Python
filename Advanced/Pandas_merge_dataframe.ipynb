```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = load_dataset("lukebarousse/data_jobs")

df = dataset["train"].to_pandas()

df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

```python
df_US_software_pivot = pd.read_csv("../Software_jobs.csv", index_col="job_posted_month")
df_US_software_pivot
```

```python
df_US = df[df["job_country"] == "United States"].copy()
```

```python
df_US["job_posted_month"] = df_US["job_posted_date"].dt.strftime("%B")
```

```python
pivot_df_US = df_US.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
```

```python
pivot_df_US.reset_index(inplace=True)

pivot_df_US["month_no"] = pd.to_datetime(pivot_df_US["job_posted_month"], format="%B").dt.month

pivot_df_US.sort_values("month_no", inplace=True)

pivot_df_US.set_index("job_posted_month", inplace=True)

pivot_df_US.drop(columns="month_no", inplace=True)
```

```python

pivot_df_US
```

```python
df_US_merged = pivot_df_US.merge(df_US_software_pivot, on="job_posted_month")
```

```python
top_5_jobs = df_US_merged.sum().sort_values(ascending=False).head().index.to_list()
top_5_jobs
```

```python
plot_us_jobs = df_US_merged[top_5_jobs].plot(kind="line")
plt.title("monthly job postings for top tech jobs in US")
plt.xlabel("2023")
plt.ylabel("job count")
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
